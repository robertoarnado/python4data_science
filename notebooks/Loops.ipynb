<h3>For Loops </h3> 


Use loops to print out the elements in the list <code>A</code>:


````python
A = [3,4,5]
for i in A:
    print(i)
````

````python
frutas = ["naranja", "manzana", "mango"]
cantidades = [10, 40, 59]
for fruta in frutas:
    print(fruta)
    print(f"la fruta con mayor exportacion es el {fruta}")
    
for cantidad in cantidades:
    print(f"se exporto {cantidad} kg de manzana en en 2020")
    
# Recorriendo una dos lista con la funcion zip

for cantidad, fruta in zip(cantidades, frutas):
    print (f"se proceso la {fruta} en total:")
    print (f"{cantidad}  kg en el mes de ocutubre" )
    
````

````python
#iterar with zip

animales = ["perro", "gato", "loro"]
for animal in animales:
    print (animal)

cantidad = [ 14, 5, 1]
for peso in cantidad:
    print (peso)

for peso, animal in zip(cantidad, animales):
    print(f" El {animal} tiene un peso de:")
    print(f"{peso} kg")
````

## Loops

### Range

Sometimes, you might want to repeat a given operation many times. Repeated executions like this are performed by <b>loops</b>. We will look at two types of loops, <code>for</code> loops and <code>while</code> loops.

Before we discuss loops lets discuss the <code>range</code> object. It is helpful to think of the range object as an ordered list. For now, let's look at the simplest case. If we would like to generate an object that contains elements ordered from 0 to 2 we simply use the following command:

### What is <code>for</code> loop?

The <code>for</code> loops enables to execute code block multiple times. For example you would use this if you would like to print out every element in a list.\
Let's try to use a <code>for</code> loop to print all the years presented in the list <code>dates</code>:


This can be done as follows:

````python
# For loop example
dates = [1990, 1995, 2000, 2005, 2010, 2015]
N = len(dates)
for i in range(N):
    print(dates[i])
````

The code in the indent is executed <code>N</code> times, each time the value of <code>i</code> is increased by 1 for every execution. The statement executed is to <code>print</code> out the value in the list at index <code>i</code> as shown here:
